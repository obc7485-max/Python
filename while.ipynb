```python
customer = "토르"
person = "Unknown"

while person != customer : # 이 while문은 'person' 변수가 'customer' 변수와 같지 않은 동안 반복됩니다.
    print("{0}, 커피가 준비 되었습니다.".format(customer))
    person = input("이름이 어떻게 되세요? ")
```
